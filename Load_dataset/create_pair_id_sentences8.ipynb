```python
import numpy as np
import pandas as pd
import json
```

```python
df_train_raw = pd.read_csv("data/train-unbalanced.csv",sep='|',header=None,names=['thread','sentences','labels'])
df_test_raw = pd.read_csv("data/test-unbalanced.csv",sep='|',header=None,names=['thread','sentences','labels'])
```

```python
def create_pair_id_sentences(df_raw,csv_name):
    res = []
    for index,row in df_raw.iterrows():
        thread = row['thread']
        if len(thread.split()) == 1:
            list_sentences = row['sentences'].split()
            list_label = row['labels'].split()
            res.append([thread,list_sentences[0],list_label[0]])
            for i in range(1,len(list_sentences)):
                res.append([list_sentences[i-1],list_sentences[i],list_label[i]])
        if index%10000==0:
            print(index,'/',len(df_raw))

    df_res = pd.DataFrame(data=res,columns=['previous_sentence','sentence','label']) 
    print("shape ", df_res.shape)
    df_res.to_csv(path_or_buf='data/'+csv_name,sep=',')
    

def create_5_id_sentences(df_raw,csv_name):
    res = []
    for index,row in df_raw.iterrows():
        thread = row['thread']
        if len(thread.split()) == 1:
            list_sentences = row['sentences'].split()
            list_label = row['labels'].split()
            res.append([-1,-1,-1,thread,list_sentences[0],list_label[0]])
            if len(list_sentences)>=2:
                res.append([-1,-1,thread,list_sentences[0],list_sentences[1],list_label[1]])
            if len(list_sentences)>=3:
                res.append([-1,thread,list_sentences[0],list_sentences[1],list_sentences[2],list_label[2]])
            if len(list_sentences)>=4:
                res.append([thread,list_sentences[0],list_sentences[1],list_sentences[2],list_sentences[3],list_label[3]])
            for i in range(4,len(list_sentences)):
                res.append([list_sentences[i-4],list_sentences[i-3],list_sentences[i-2],list_sentences[i-1],list_sentences[i],list_label[i]])
        if index%10000==0:
            print(index,'/',len(df_raw))

    df_res = pd.DataFrame(data=res,columns=['previous4_sentence','previous3_sentence','previous2_sentence','previous_sentence','sentence','label']) 
    print("shape ", df_res.shape)
    df_res.to_csv(path_or_buf='data/'+csv_name,sep=',')

def create_8_id_sentences(df_raw,csv_name):
    res = []
    for index,row in df_raw.iterrows():
        thread = row['thread']
        if len(thread.split()) == 1:
            list_sentences = row['sentences'].split()
            list_label = row['labels'].split()
            res.append([-1,-1,-1,-1,-1,-1,-1,thread,list_sentences[0],list_label[0]])
            if len(list_sentences)>=2:
                res.append([-1,-1,-1,-1,-1,-1,thread,list_sentences[0],list_sentences[1],list_label[1]])
            if len(list_sentences)>=3:
                res.append([-1,-1,-1,-1,-1,thread,list_sentences[0],list_sentences[1],list_sentences[2],list_label[2]])
            if len(list_sentences)>=4:
                res.append([-1,-1,-1,-1,thread,list_sentences[0],list_sentences[1],list_sentences[2],list_sentences[3],list_label[3]])
            if len(list_sentences)>=5:
                res.append([-1,-1,-1,thread,list_sentences[0],list_sentences[1],list_sentences[2],list_sentences[3],list_sentences[4],list_label[4]])
            if len(list_sentences)>=6:
                res.append([-1,-1,thread,list_sentences[0],list_sentences[1],list_sentences[2],list_sentences[3],list_sentences[4],list_sentences[5],list_label[5]])
            if len(list_sentences)>=7:
                res.append([-1,thread,list_sentences[0],list_sentences[1],list_sentences[2],list_sentences[3],list_sentences[4],list_sentences[5],list_sentences[6],list_label[6]])
            if len(list_sentences)>=8:
                res.append([thread,list_sentences[0],list_sentences[1],list_sentences[2],list_sentences[3],list_sentences[4],list_sentences[5],list_sentences[6],list_sentences[7],list_label[7]])
            for i in range(8,len(list_sentences)):
                res.append([list_sentences[i-8],list_sentences[i-7],list_sentences[i-6],list_sentences[i-5],list_sentences[i-4],list_sentences[i-3],list_sentences[i-2],list_sentences[i-1],list_sentences[i],list_label[i]])
        if index%10000==0:
            print(index,'/',len(df_raw))

    df_res = pd.DataFrame(data=res,columns=['prev8','prev7','prev6','prev5','prev4','prev3','prev2','prev1','sentence','label']) 
    print("shape ", df_res.shape)
    df_res.to_csv(path_or_buf='data/'+csv_name,sep=',')

```

```python
create_8_id_sentences(df_train_raw,'8_id_train.csv')
create_8_id_sentences(df_test_raw,'8_id_test.csv')
```
